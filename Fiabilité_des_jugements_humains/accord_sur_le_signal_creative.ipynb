```python
# On regroupe
grouped = df_mm_message.groupby(['msg_index', 'conversation_pair_id'])['creative']

# On filtre : on ne garde que les groupes où le nombre d'entrées (count) est > 2
groupes_plus_de_2 = grouped.agg(list)[grouped.count() > 2]

print(groupes_plus_de_2)
```

```python
#calcul du taux d'accord brut sur le label creative pour les paires où au moins deux utilisateurs ont réagi au même message
liste_taux_accord_brut=[]
for list_creative in df_mm_message.groupby(['msg_index','conversation_pair_id']).agg(list)["creative"]:
    taux_actuel=0
    for i in range(len(list_creative)-1):
        for j in range(i+1,len(list_creative)):
            if list_creative[i]==list_creative[j]:
                taux_actuel+=1
    taux_actuel=taux_actuel/(len(list_creative)*(len(list_creative)-1)/2)
    liste_taux_accord_brut.append(taux_actuel)
print(liste_taux_accord_brut)
```

On a que des groupes de 2, donc le taux d'accord brut est soit 0 (si les deux utilisateurs n'ont pas réagi de la même manière) soit 1 (si les deux utilisateurs ont réagi de la même manière). Nous allons calculer la moyenne de ces taux d'accord brut pour avoir une idée de l'accord global sur le signal "creative".

```python
len(liste_taux_accord_brut)
```

```python
moyenne_taux_accord_brut=sum(liste_taux_accord_brut)/len(liste_taux_accord_brut)
print(moyenne_taux_accord_brut)
```

On voit que le score est très élevé, ce qui signifie que les utilisateurs sont globalement d'accord sur le signal "creative". ### 11.	Calculez le κ de Cohen pour les conversations avec réactions multiples. Interprétez le résultat au regard des qualificatifs useful et complete (comparez leurs κ respectifs).

```python
# trouver les conversations avec des réactions multiples
df_mm_conversation=df_comparia_reactions[df_comparia_reactions.duplicated(subset=['conversation_pair_id'], keep=False)]
df_mm_conversation.head(10)
```

```python
df_mm_conversation.groupby(['conversation_pair_id']).agg(list)
```

```python
from sklearn.metrics import cohen_kappa_score
from itertools import combinations

def compute_kappa(df, label, id_col='conversation_pair_id'):
    """
    Calcule le κ de Cohen pour un signal donné.

    Pour chaque conversation avec plusieurs réactions, extrait toutes les
    paires d'annotateurs (via combinaisons) et calcule le kappa global.

    Paramètres
    ----------
    df    : DataFrame contenant les réactions multiples (déjà filtré)
    label : nom de la colonne du signal ('creative', 'useful', 'complete')
    id_col: colonne identifiant les conversations (default: 'conversation_pair_id')

    Retourne
    --------
    float : valeur du κ de Cohen
    """
    ratings_a, ratings_b = [], []

    for _, group in df.groupby(id_col):
        labels = group[label].dropna().tolist()
        for a, b in combinations(labels, 2):
            ratings_a.append(a)
            ratings_b.append(b)

    if len(ratings_a) < 2:
        return float('nan')

    return cohen_kappa_score(ratings_a, ratings_b)

# Calcul du kappa pour chaque signal
k_cohen_creative  = compute_kappa(df_mm_conversation, 'creative')
k_cohen_useful    = compute_kappa(df_mm_conversation, 'useful')
k_cohen_complete  = compute_kappa(df_mm_conversation, 'complete')

print(f"κ de Cohen — creative  : {k_cohen_creative:.4f}")
print(f"κ de Cohen — useful    : {k_cohen_useful:.4f}")
print(f"κ de Cohen — complete  : {k_cohen_complete:.4f}")

```

### Interprétation

| Signal | κ de Cohen | Qualificatif (Landis & Koch) | Interprétation |
|--------|-----------|------------------------------|----------------|
| **useful** | 0.36 | Accord passable (0.21–0.40) | L'utilité est un critère relativement objectif ("la réponse répond-elle à la question ?"), d'où un accord entre annotateurs |
| **complete** | 0.29 | Accord passable (0.21–0.40) | La complétude a le kappa de Cohen le plus bas mais présentant un accord passable tout de même |
| **creative** | 0.31| Accord passable (0.21–0.40) | Un κ représentant un accord moyen mais il ne faut pas oublier que plusieurs annotateurs ont voté sur la même conversation cassant l'accord |



```python
nb_visitor_id_grouped = df_mm_conversation.groupby(['conversation_pair_id'])['visitor_id'].nunique()
print(nb_visitor_id_grouped.value_counts())
```

```python
#calculer erreur standard du kappa de Cohen pour le label creative
import numpy as np
def cohen_kappa_std_error(kappa, n):
    """
    Calcule l'erreur standard du kappa de Cohen.
    kappa : valeur du kappa de Cohen
    n : nombre total d'annotations (paires d'annotateurs)
    Hypothese : distribution binaire equilibree (pe = 0.5)
    """
    if n == 0:
        return float('nan')
    p0 = (kappa + 1) / 2  # Proportion d'accord observe
    pe = 0.5               # Proportion d'accord attendu (hypothese equilibree)
    var_kappa = p0 * (1 - p0) / (n * (1 - pe) ** 2)  # Formule de Fleiss
    return np.sqrt(var_kappa)
n_annotations = len(nb_visitor_id_grouped)
std_error_kappa_creative = cohen_kappa_std_error(k_cohen_creative, n_annotations)
print(f"Erreur standard du kappa de Cohen (creative) : {std_error_kappa_creative:.4f}")
```

## 12.	Formulez deux hypothèses sur les raisons pour lesquelles le signal creative est probablement plus bruité que le signal useful.

```python
df_mm_conversation.groupby(['conversation_pair_id']).agg(list)["creative"]
```

```python
df_mm_conversation.groupby(['conversation_pair_id']).agg(list)["useful"]
```

### Réponse

**Hypothèse 1 — Subjectivité inhérente du concept de créativité.**  

La créativité est une notion intrinsèquement subjective. Contrairement à *useful* ("la réponse a-t-elle aidé l'utilisateur ?") qui peut se vérifier factuellement, juger si une réponse est "créative" dépend des attentes esthétiques, du parcours culturel et des goûts personnels de l'annotateur. Deux annotateurs peuvent donc être en total désaccord tout en étant chacun cohérent avec sa propre définition interne de la créativité.

**Hypothèse 2 : L'applicabilité variable selon le type de requête (Context Mismatch).**

Le critère useful est universel : quel que soit le prompt (générer du code, répondre à une question d'histoire, écrire un poème), on peut toujours évaluer si la réponse a été utile à l'utilisateur.

À l'inverse, le critère creative n'est pas pertinent pour toutes les requêtes. Si un utilisateur demande "Comment calculer la TVA ?" ou "Corrige ce code Python", la réponse attendue doit être factuelle, pas créative.
