On voit que l'on a 2 messages identique dansle dataframe qui nous permettent de calculer le taux d'accord brut sur le label creative pour les paires où au moins deux utilisateurs ont réagi au même message. Le problème est que pour chaque paire l'utilisateur est le même donc le même utilisateur a réagi au même message deux fois.

```python
df_mm_message.groupby(['msg_index','conversation_pair_id']).agg(list)["creative"]
```

```python
#calcul du taux d'accord brut sur le label creative pour les paires où au moins deux utilisateurs ont réagi au même message
liste_taux_accord_brut=[]
for list_creative in df_mm_message.groupby(['msg_index','conversation_pair_id']).agg(list)["creative"]:
    taux_actuel=0
    for i in range(len(list_creative)-1):
        for j in range(i+1,len(list_creative)):
            if list_creative[i]==list_creative[j]:
                taux_actuel+=1
    taux_actuel=taux_actuel/(len(list_creative)*(len(list_creative)-1)/2)
    liste_taux_accord_brut.append(taux_actuel)
print(liste_taux_accord_brut)
```

Bon comme c'est la même personne le taux d'accord brut est de 100% pour les 2 messages.

### 11.	Calculez le κ de Cohen pour les conversations avec réactions multiples. Interprétez le résultat au regard des qualificatifs useful et complete (comparez leurs κ respectifs).

```python
# trouver les conversations avec des réactions multiples
df_mm_conversation=df_comparia_reactions[df_comparia_reactions.duplicated(subset=['conversation_pair_id'], keep=False)]
df_mm_conversation.head(10)
```

```python
df_mm_conversation.groupby(['conversation_pair_id']).agg(list)
```

```python
from sklearn.metrics import cohen_kappa_score
from itertools import combinations

def compute_cohen_kappa(df_grouped, label):
    """
    Calcule le kappa de Cohen pour un label donné,
    sur toutes les paires d'annotateurs au sein de chaque conversation.
    """
    annotator1 = []
    annotator2 = []
    for values in df_grouped[label]:
        # Pour chaque conversation, on forme toutes les paires possibles
        for v1, v2 in combinations(values, 2):
            annotator1.append(v1)
            annotator2.append(v2)
    if len(annotator1) < 2:
        return float('nan')
    return cohen_kappa_score(annotator1, annotator2)

grouped = df_mm_conversation.groupby(['conversation_pair_id']).agg(list)

k_cohen_creative  = compute_cohen_kappa(grouped, 'creative')
k_cohen_useful    = compute_cohen_kappa(grouped, 'useful')
k_cohen_complete  = compute_cohen_kappa(grouped, 'complete')

print(f"κ de Cohen (creative)  : {k_cohen_creative:.4f}")
print(f"κ de Cohen (useful)    : {k_cohen_useful:.4f}")
print(f"κ de Cohen (complete)  : {k_cohen_complete:.4f}")
```

### Interprétation

| Signal | κ de Cohen | Qualificatif (Landis & Koch) | Interprétation |
|--------|-----------|------------------------------|----------------|
| **useful** | 0.3913 | Accord passable (0.21–0.40) | L'utilité est un critère relativement objectif ("la réponse répond-elle à la question ?"), d'où un accord raisonnable entre annotateurs |
| **complete** | 0.4324 | Accord modéré (0.41–0.60) | La complétude est également assez factuelle ("manque-t-il des éléments ?"), ce qui produit le meilleur accord des trois |
| **creative** | −0.1053 | Désaccord (< 0) | Un κ négatif signifie que les annotateurs sont **moins d'accord que le hasard** : ils ont des conceptions opposées de la créativité |

**Constat clé** : `useful` (0.39) et `complete` (0.43) montrent un accord passable à modéré, tandis que `creative` (−0.11) révèle un véritable désaccord systématique. Le signal *creative* est donc **beaucoup plus bruité** que les deux autres — il ne peut pas être utilisé de manière fiable sans redéfinition plus précise du critère ou formation des annotateurs.

```python
nb_visitor_id_grouped = df_mm_conversation.groupby(['conversation_pair_id'])['visitor_id'].nunique()
print(nb_visitor_id_grouped.value_counts())
```

```python
#calculer erreur standard du kappa de Cohen pour le label creative
import numpy as np
def cohen_kappa_std_error(kappa, n):
    """
    Calcule l'erreur standard du kappa de Cohen.
    kappa : valeur du kappa de Cohen
    n : nombre total d'annotations (paires d'annotateurs)
    """
    if n == 0:
        return float('nan')
    p0 = (kappa + 1) / 2  # Proportion d'accord observé
    pe = (1 - p0) * (1 - p0) + p0 * p0  # Proportion d'accord attendu
    var_kappa = (p0 * (1 - p0) + pe * (1 - pe)) / n  # Variance du kappa
    return np.sqrt(var_kappa)
n_annotations = len(nb_visitor_id_grouped)
std_error_kappa_creative = cohen_kappa_std_error(k_cohen_creative, n_annotations)
print(f"Erreur standard du κ de Cohen (creative) : {std_error_kappa_creative:.4f}")
```

## 12.	Formulez deux hypothèses sur les raisons pour lesquelles le signal creative est probablement plus bruité que le signal useful.

```python
df_mm_conversation.groupby(['conversation_pair_id']).agg(list)["creative"]
```

```python
df_mm_conversation.groupby(['conversation_pair_id']).agg(list)["useful"]
```

### Réponse

**Hypothèse 1 — Subjectivité inhérente du concept de créativité.**  
La créativité est une notion intrinsèquement subjective. Contrairement à *useful* ("la réponse a-t-elle aidé l'utilisateur ?") qui peut se vérifier factuellement, juger si une réponse est "créative" dépend des attentes esthétiques, du parcours culturel et des goûts personnels de l'annotateur. Deux annotateurs peuvent donc être en total désaccord tout en étant chacun cohérent avec sa propre définition interne de la créativité.

**Hypothèse 2 — Faible prévalence et sensibilité au contexte conversationnel.**  
Le signal *creative* est plus rare car le dataset est un dataset de conversation entre utilisateur et llm et il arrive plus souvent d'avoir des réponse utile que créative. Donc on a un gros déséquilibre et donc un désaccord pèse plus lourd sur la formule du k qui est:
$$\kappa = \frac{p_o - p_e}{1 - p_e}$$
où $p_o$ est la proportion d'accord observé et $p_e$ la proportion d'accord attendu par hasard. Avec un signal rare, même un petit désaccord peut faire chuter $p_o$ de manière significative, tandis que $p_e$ reste relativement élevé, ce qui entraîne un κ négatif. De plus, la créativité peut être perçue différemment selon le contexte de la conversation, rendant l'accord encore plus difficile à atteindre.
